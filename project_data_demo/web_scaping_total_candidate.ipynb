```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import time

# Setup Chrome driver
driver = webdriver.Chrome()  # Make sure chromedriver is in PATH

all_data = []

# Loop through pages (example: pages 1 to 2, update range as needed)
for page in range(1, 42):
    url = f'https://www.myneta.info/Maharashtra2024/index.php?action=summary&subAction=candidates_analyzed&sort=candidate&page={page}'
    #https://www.myneta.info/Maharashtra2024/index.php?action=summary&subAction=candidates_analyzed&sort=candidate&page=1
    driver.get(url)
    time.sleep(3)  # wait for page to load JS
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    table = soup.find('table', attrs={'class': 'w3-table w3-bordered'})
    
    # Read rows
    rows = table.find_all('tr')
    headers = [th.text.strip() for th in rows[0].find_all('th')]
    
    for row in rows[1:]:
        cols = [td.text.strip() for td in row.find_all('td')]
        if cols:
            all_data.append(cols)

# Save to DataFrame
df = pd.DataFrame(all_data, columns=headers)
df.to_csv('maharashtra_assembly_candidates.csv', index=False)
print(df.head())

driver.quit()
```
