```python
import re

def clean_asset_value(value):
    """Extract numeric value (remove Rs, commas, words)."""
    try:
        # Example: "Rs 8,00,000 ~ 8 Lacs+" → "800000"
        num = re.findall(r'[\d,]+', value)
        if num:
            return int(num[0].replace(",", ""))  # remove commas
        else:
            return 0
    except:
        return 0

def clean_asset_label(value):
    """Extract human readable part after ~ sign."""
    try:
        if "~" in value:
            return value.split("~")[1].strip()
        else:
            return value
    except:
        return value
```

```python
# For Total Assets
total_candidates_df["Total_Assets_Value"] = total_candidates_df["Total_Assets"].apply(clean_asset_value)
total_candidates_df["Total_Assets_Label"] = total_candidates_df["Total_Assets"].apply(clean_asset_label)

# For Liabilities
total_candidates_df["Liabilities_Value"] = total_candidates_df["Liabilities"].apply(clean_asset_value)
total_candidates_df["Liabilities_Label"] = total_candidates_df["Liabilities"].apply(clean_asset_label)
```

```python
def number_to_words(num):
    """Convert number into Indian numbering system words (Lakh, Crore)."""
    if num >= 1_00_00_000:   # Crores
        return f"{num/1_00_00_000:.2f} Crore+"
    elif num >= 1_00_000:    # Lakhs
        return f"{num/1_00_000:.2f} Lakh+"
    elif num >= 1_000:       # Thousands
        return f"{num/1_000:.2f} Thousand+"
    elif num > 0:
        return str(num)
    else:
        return "0"
```

```python
total_candidates_df["Net_Worth"] = (
    total_candidates_df["Total_Assets_Value"] - total_candidates_df["Liabilities_Value"]
)

total_candidates_df["Net_Worth_Words"] = total_candidates_df["Net_Worth"].apply(number_to_words)
```
